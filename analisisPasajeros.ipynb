```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/anaArango365/Seminario/refs/heads/main/salida_mensual_pasajeros_aeropuerto_destino_internacional.csv")

# Renombrar columnas
df = df.rename(columns={
    "sal_codigo": "codigo_pais",
    "sal_destinoint": "pais_destino",
    "sal_indicador": "indicador_salida",
    "sal_periodo": "periodo",
    "sal_valor": "valor_salida"
})

df["periodo"] = pd.to_datetime(df["periodo"], format="%Y%m")

tendencia = df.groupby("periodo")["valor_salida"].sum().reset_index()

antes_2020 = tendencia[tendencia["periodo"] < "2020-01-01"]
desde_2020 = tendencia[tendencia["periodo"] >= "2020-01-01"]

plt.figure(figsize=(18, 6))

plt.bar(antes_2020["periodo"], antes_2020["valor_salida"], color='lightgreen', width=20, label='Antes de 2020')

plt.bar(desde_2020["periodo"], desde_2020["valor_salida"], color='lightcoral', width=20, label='Desde 2020')

plt.plot(tendencia["periodo"], tendencia["valor_salida"], color='steelblue', marker='o', linestyle='-', label='Tendencia')

plt.title("Tendencia cronológica de viajes internacionales (pasajeros mensuales)")
plt.xlabel("Periodo")
plt.ylabel("Total de pasajeros")
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.legend()
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Renombrar columnas
df = df.rename(columns={
    "sal_codigo": "codigo_pais",
    "sal_destinoint": "pais_destino",
    "sal_indicador": "indicador_salida",
    "sal_periodo": "periodo",
    "sal_valor": "valor_salida"
})

# Convertir periodo a fecha
df["periodo"] = pd.to_datetime(df["periodo"], format="%Y%m")

# Obtener los 6 países más visitados por total
top6_paises = df.groupby("pais_destino")["valor_salida"].sum().sort_values(ascending=False).head(6).index

# Crear subgráficas 2x3
fig, axs = plt.subplots(2, 3, figsize=(20, 10), sharex=True, sharey=True)
axs = axs.flatten()

# Para cada país, generar la gráfica
for i, pais in enumerate(top6_paises):
    df_pais = df[df["pais_destino"] == pais]

    # Agrupar por periodo
    tendencia = df_pais.groupby("periodo")["valor_salida"].sum().reset_index()

    # Dividir antes y después de 2020
    antes_2020 = tendencia[tendencia["periodo"] < "2020-01-01"]
    desde_2020 = tendencia[tendencia["periodo"] >= "2020-01-01"]

    # Gráfico
    axs[i].bar(antes_2020["periodo"], antes_2020["valor_salida"], color='lightgreen', width=20, label='Antes de 2020')
    axs[i].bar(desde_2020["periodo"], desde_2020["valor_salida"], color='lightcoral', width=20, label='Desde 2020')
    axs[i].plot(tendencia["periodo"], tendencia["valor_salida"], color='steelblue', marker='o', linestyle='-', linewidth=2.2, label='Tendencia')

    axs[i].set_title(f"País: {pais}", fontsize=13, fontweight='bold')
    axs[i].tick_params(axis='x', rotation=45)
    axs[i].grid(True, axis='y', linestyle='--', alpha=0.5)

# Título general
plt.suptitle("Tendencia cronológica de viajes internacionales a los 6 países más visitados", fontsize=18, weight='bold')
plt.tight_layout(rect=[0, 0.04, 1, 0.95])
plt.legend(loc='upper right')
plt.show()
```

```python

# Renombrar columnas
df = df.rename(columns={
    "sal_codigo": "codigo_pais",
    "sal_destinoint": "pais_destino",
    "sal_indicador": "indicador_salida",
    "sal_periodo": "periodo",
    "sal_valor": "valor_salida"
})

# Convertir periodo a datetime
df["periodo"] = pd.to_datetime(df["periodo"], format="%Y%m")

# Obtener los 6 países más visitados
top6_paises = df.groupby("pais_destino")["valor_salida"].sum().sort_values(ascending=False).head(6).index.tolist()

# Filtrar solo esos países
df_top6 = df[df["pais_destino"].isin(top6_paises)]

# Agrupar por periodo y país
tendencia = df_top6.groupby(["periodo", "pais_destino"])["valor_salida"].sum().reset_index()

# Pivotear para tener cada país como una columna
pivot = tendencia.pivot(index="periodo", columns="pais_destino", values="valor_salida").fillna(0)

# Crear gráfica
plt.figure(figsize=(18, 7))

# Colores personalizados (puedes cambiar)
colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#17becf']

# Graficar cada país
for i, pais in enumerate(pivot.columns):
    plt.plot(pivot.index, pivot[pais], label=pais, linewidth=2.5, marker='o', markersize=4, color=colores[i])

# Estilo
plt.title("Comparativo de los 6 países más visitados (valor mensual de salida)", fontsize=16, fontweight='bold')
plt.xlabel("Periodo")
plt.ylabel("Pasajeros (valor de salida)")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend(title="País destino", loc='upper left')
plt.tight_layout()

plt.show()
```

```python

# Renombrar columnas
df = df.rename(columns={
    "sal_codigo": "codigo_pais",
    "sal_destinoint": "pais_destino",
    "sal_indicador": "indicador_salida",
    "sal_periodo": "periodo",
    "sal_valor": "valor_salida"
})

# Convertir periodo a fecha
df["periodo"] = pd.to_datetime(df["periodo"], format="%Y%m")

# Obtener el país más visitado
pais_top = df.groupby("pais_destino")["valor_salida"].sum().sort_values(ascending=False).idxmax()
df_top = df[df["pais_destino"] == pais_top]

# Agrupar por periodo
tendencia = df_top.groupby("periodo")["valor_salida"].sum().reset_index()

# Dividir antes y después de 2020
antes_2020 = tendencia[tendencia["periodo"] < "2020-01-01"]
desde_2020 = tendencia[tendencia["periodo"] >= "2020-01-01"]

# Crear gráfica
plt.figure(figsize=(18, 6))

# Barras
plt.bar(antes_2020["periodo"], antes_2020["valor_salida"], color='lightgreen', width=20, label='Antes de 2020')
plt.bar(desde_2020["periodo"], desde_2020["valor_salida"], color='lightcoral', width=20, label='Desde 2020')

# Línea de tendencia
plt.plot(tendencia["periodo"], tendencia["valor_salida"], color='steelblue', marker='o', linewidth=2.5, label='Tendencia')

# Títulos y etiquetas
plt.title(f"Tendencia de viajes internacionales a {pais_top}", fontsize=16, fontweight='bold')
plt.xlabel("Periodo")
plt.ylabel("Total de pasajeros")
plt.xticks(rotation=45)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()

plt.show()
```
